Hệ thống đề xuất dựa trên bộ lọc cộng tác


```python
x = ratings_with_name.groupby('ten_kh').count()['diem_dg'] >= 0
padhe_likhe_users = x[x].index
x.sum()
# padhe_likhe_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['ten_kh'].isin(
    padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('ten_mon_an').count()['diem_dg'] > 0
famous_books = y[y].index
int(y.sum())
```

```python
final_ratings = filtered_rating[filtered_rating['ten_mon_an'].isin(
    famous_books)]
```

```python
# final_ratings['ten_kh']
```

```python
pt = final_ratings.pivot_table(
    index='ten_kh', columns='ten_mon_an', values='diem_dg')
```

```python
data= pt
data
```

```python
item = list(data.columns)
```

```python
user = list(data.index)
```

```python
data_matrix = data[item[0:]].fillna(0).values
# data_matrix
```

```python
Sim = cosine_similarity(data_matrix, data_matrix)
Sim = np.round(Sim, decimals=2)
# Sim
```

```python
pd.DataFrame(Sim, columns=user, index=user)
```

```python
def predict(A, u, i, k=2):
    user_rated_i = np.where(A[:, i] != 0)[0]
    sim = Sim[u, user_rated_i]
    a = (np.argsort(sim))[-k:]
    nearest_s = sim[a]
    rating = A[user_rated_i[a], i]
    r_bar = (rating*nearest_s).sum()/(np.abs(nearest_s).sum())
    return np.round(r_bar, 2)
```

```python
predict_matrix = data_matrix.copy()
for u in range(0, predict_matrix.shape[0]):
  for i in range(0, predict_matrix.shape[1]):
    if data_matrix[u, i] == 0:
      id = i
      predict_matrix[u, i] = predict(data_matrix, u, id)
# predict_matrix
```

```python
pd.DataFrame(predict_matrix, columns=item[0:], index=user)
```
